## Evaluate Model's Performance

#### Load test dataset

```python
test_dataset = "../Dataset/yolo_training/florida_five_anole_10000_v4/test_data.yaml"
```

#### Validate YOLOv12x

```python
model = YOLO("./runs/detect/train22_yolov12x_dataset_v4/weights/best.pt")

metrics = model.val(data=test_dataset)  # no arguments needed, dataset and settings remembered

print("Mean average precision at IoU=0.50:", metrics.box.map50)
print("Mean average precision at IoU=0.75:", metrics.box.map75)  # mAP75
print("Mean average precision at IoU=0.5-0.95:", metrics.box.maps)  # a list contains map50-95 of each category

print("precision:", metrics.results_dict['metrics/precision(B)'])
print("recall:", metrics.results_dict['metrics/recall(B)']) 
print("F1 score:", metrics.box.f1)
print("speed:", metrics.speed)
```

#### Validate YOLOv12x
Conf: 0.3
IoU: 0.2

```python
model = YOLO("./runs/detect/train22_yolov12x_dataset_v4/weights/best.pt")

metrics = model.val(data=test_dataset, conf=0.3, iou=0.2, max_det=5)  # no arguments needed, dataset and settings remembered

print("Mean average precision at IoU=0.50:", metrics.box.map50)
print("Mean average precision at IoU=0.75:", metrics.box.map75)  # mAP75
print("Mean average precision at IoU=0.5-0.95:", metrics.box.maps)  # a list contains map50-95 of each category

print("precision:", metrics.results_dict['metrics/precision(B)'])
print("recall:", metrics.results_dict['metrics/recall(B)']) 
print("F1 score:", metrics.box.f1)
print("speed:", metrics.speed)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
classes = ['Overall', 'bark_anole', 'brown_anole', 'created_anole', 'green_anole', 'knight_anole']

# Metrics for each model
yolov12x = [0.874, 0.907, 0.857, 0.861, 0.890, 0.854]  # F1 scores
pipeline = [0.921, 0.88, 0.915, 0.8601, 0.944, 0.917]  # F1 scores
yolov8x = [0.890, 0.924, 0.862, 0.880, 0.915, 0.869]   # F1 scores

# Create plot
plt.figure(figsize=(12, 6))
bar_width = 0.25
x_pos = np.arange(len(classes))

plt.bar(x_pos - bar_width, yolov12x, bar_width, label='YOLOv12x', alpha=0.8, color='skyblue')
plt.bar(x_pos, pipeline, bar_width, label='Pipeline', alpha=0.8, color='lightcoral')
plt.bar(x_pos + bar_width, yolov8x, bar_width, label='YOLOv8x', alpha=0.8, color='lightgreen')

plt.title('F1-Score Comparison Across Models', fontsize=14, fontweight='bold')
plt.ylabel('F1-Score')
plt.xlabel('Classes')
plt.xticks(x_pos, classes, rotation=45)
plt.legend()
plt.grid(True, alpha=0.3)

# Add value labels on bars
for i, v in enumerate(yolov12x):
    plt.text(i - bar_width, v + 0.005, f'{v:.3f}', ha='center', fontsize=8)
for i, v in enumerate(pipeline):
    plt.text(i, v + 0.005, f'{v:.3f}', ha='center', fontsize=8)
for i, v in enumerate(yolov8x):
    plt.text(i + bar_width, v + 0.005, f'{v:.3f}', ha='center', fontsize=8)

plt.tight_layout()
plt.show()
```
