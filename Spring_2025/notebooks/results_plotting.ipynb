## Results Plot

```python
import matplotlib.pyplot as plt
import numpy as np

# Data from the table
classes = ['Overall', 'bark_anole', 'brown_anole', 'created_anole', 'green_anole', 'knight_anole']

# YOLOv12x metrics
yolov12x_precision = [0.736, 0.78, 0.73, 0.69, 0.76, 0.72]
yolov12x_recall = [0.720, 0.71, 0.63, 0.74, 0.72, 0.80]
yolov12x_f1 = [0.724, 0.74, 0.68, 0.71, 0.74, 0.75]

# Pipeline metrics
pipeline_precision = [0.882, 0.93, 0.85, 0.87, 0.89, 0.87]
pipeline_recall = [0.896, 0.93, 0.85, 0.92, 0.89, 0.89]
pipeline_f1 = [0.890, 0.93, 0.85, 0.90, 0.89, 0.88]

# YOLOv8x metrics
yolov8x_precision = [0.798, 0.79, 0.82, 0.81, 0.81, 0.76]
yolov8x_recall = [0.806, 0.80, 0.74, 0.83, 0.82, 0.84]
yolov8x_f1 = [0.802, 0.79, 0.78, 0.82, 0.82, 0.80]

# Plot 4: Overall metrics comparison
overall_metrics = ['Precision', 'Recall', 'F1-Score']
yolov12x_overall = [0.736, 0.720, 0.724]
pipeline_overall = [0.882, 0.896, 0.890]
yolov8x_overall = [0.798, 0.806, 0.802]
```

```python
# Set up the plot
fig, axes = plt.subplots(2, 2, figsize=(20, 12))
fig.suptitle('Model Performance Comparison: Precision, Recall, and F1-Score', fontsize=16, fontweight='bold')

# Width of bars
bar_width = 0.25
x_pos = np.arange(len(classes))

# Plot 1: Precision comparison
axes[0, 0].bar(x_pos - bar_width, yolov12x_precision, bar_width, label='YOLOv12x', alpha=0.8, color='skyblue')
axes[0, 0].bar(x_pos, pipeline_precision, bar_width, label='Pipeline', alpha=0.8, color='lightcoral')
axes[0, 0].bar(x_pos + bar_width, yolov8x_precision, bar_width, label='YOLOv8x', alpha=0.8, color='lightgreen')
axes[0, 0].set_title('Precision Comparison', fontweight='bold')
axes[0, 0].set_ylabel('Precision')
axes[0, 0].set_xticks(x_pos)
axes[0, 0].set_xticklabels(classes, rotation=45)
axes[0, 0].legend(loc='lower right')
axes[0, 0].grid(True, alpha=0.3)

# Plot 2: Recall comparison
axes[0, 1].bar(x_pos - bar_width, yolov12x_recall, bar_width, label='YOLOv12x', alpha=0.8, color='skyblue')
axes[0, 1].bar(x_pos, pipeline_recall, bar_width, label='Pipeline', alpha=0.8, color='lightcoral')
axes[0, 1].bar(x_pos + bar_width, yolov8x_recall, bar_width, label='YOLOv8x', alpha=0.8, color='lightgreen')
axes[0, 1].set_title('Recall Comparison', fontweight='bold')
axes[0, 1].set_ylabel('Recall')
axes[0, 1].set_xticks(x_pos)
axes[0, 1].set_xticklabels(classes, rotation=45)
axes[0, 1].legend(loc='lower right')
axes[0, 1].grid(True, alpha=0.3)

# Plot 3: F1-Score comparison
axes[1, 0].bar(x_pos - bar_width, yolov12x_f1, bar_width, label='YOLOv12x', alpha=0.8, color='skyblue')
axes[1, 0].bar(x_pos, pipeline_f1, bar_width, label='Pipeline', alpha=0.8, color='lightcoral')
axes[1, 0].bar(x_pos + bar_width, yolov8x_f1, bar_width, label='YOLOv8x', alpha=0.8, color='lightgreen')
axes[1, 0].set_title('F1-Score Comparison', fontweight='bold')
axes[1, 0].set_ylabel('F1-Score')
axes[1, 0].set_xlabel('Classes')
axes[1, 0].set_xticks(x_pos)
axes[1, 0].set_xticklabels(classes, rotation=45)
axes[1, 0].legend(loc='lower right')
axes[1, 0].grid(True, alpha=0.3)

x_pos_overall = np.arange(len(overall_metrics))
axes[1, 1].bar(x_pos_overall - bar_width, yolov12x_overall, bar_width, label='YOLOv12x', alpha=0.8, color='skyblue')
axes[1, 1].bar(x_pos_overall, pipeline_overall, bar_width, label='Pipeline', alpha=0.8, color='lightcoral')
axes[1, 1].bar(x_pos_overall + bar_width, yolov8x_overall, bar_width, label='YOLOv8x', alpha=0.8, color='lightgreen')
axes[1, 1].set_title('Overall Metrics Comparison', fontweight='bold')
axes[1, 1].set_ylabel('Score')
axes[1, 1].set_xlabel('Metrics')
axes[1, 1].set_xticks(x_pos_overall)
axes[1, 1].set_xticklabels(overall_metrics)
axes[1, 1].legend(loc='lower right')
axes[1, 1].grid(True, alpha=0.3)

# Add value labels on bars
for ax in axes.flat:
    for container in ax.containers:
        ax.bar_label(container, fmt='%.3f', padding=3, fontsize=8)

plt.tight_layout()
plt.show()
```

### Precision

```python
plt.figure(figsize=(10, 6))

plt.bar(x_pos - bar_width, yolov12x_precision, bar_width,
        label='YOLOv12x', alpha=0.8, color='skyblue')
plt.bar(x_pos, pipeline_precision, bar_width,
        label='Pipeline', alpha=0.8, color='lightcoral')
plt.bar(x_pos + bar_width, yolov8x_precision, bar_width,
        label='YOLOv8x', alpha=0.8, color='lightgreen')

plt.title('Precision Comparison', fontweight='bold')
plt.ylabel('Precision')
plt.xticks(x_pos, classes, rotation=45)
plt.legend(loc='lower right')
plt.grid(True, alpha=0.3)

for container in plt.gca().containers:
    plt.bar_label(container, fmt='%.3f', padding=3, fontsize=8)

plt.tight_layout()
plt.show()
```

### Recall

```python
plt.figure(figsize=(10, 6))

plt.bar(x_pos - bar_width, yolov12x_recall, bar_width,
        label='YOLOv12x', alpha=0.8, color='skyblue')
plt.bar(x_pos, pipeline_recall, bar_width,
        label='Pipeline', alpha=0.8, color='lightcoral')
plt.bar(x_pos + bar_width, yolov8x_recall, bar_width,
        label='YOLOv8x', alpha=0.8, color='lightgreen')

plt.title('Recall Comparison', fontweight='bold')
plt.ylabel('Recall')
plt.xticks(x_pos, classes, rotation=45)
plt.legend(loc='lower right')
plt.grid(True, alpha=0.3)

for container in plt.gca().containers:
    plt.bar_label(container, fmt='%.3f', padding=3, fontsize=8)

plt.tight_layout()
plt.show()
```

### F1-score

```python
plt.figure(figsize=(10, 6))

plt.bar(x_pos - bar_width, yolov12x_f1, bar_width,
        label='YOLOv12x', alpha=0.8, color='skyblue')
plt.bar(x_pos, pipeline_f1, bar_width,
        label='Pipeline', alpha=0.8, color='lightcoral')
plt.bar(x_pos + bar_width, yolov8x_f1, bar_width,
        label='YOLOv8x', alpha=0.8, color='lightgreen')

plt.title('F1-score Comparison', fontweight='bold')
plt.ylabel('F1-score')
plt.xlabel('Classes')
plt.xticks(x_pos, classes, rotation=45)
plt.legend(loc='lower right')
plt.grid(True, alpha=0.3)

for container in plt.gca().containers:
    plt.bar_label(container, fmt='%.3f', padding=3, fontsize=8)

plt.tight_layout()
plt.show()
```

### Overall Metrics

```python
plt.figure(figsize=(8, 6))

plt.bar(x_pos_overall - bar_width, yolov12x_overall, bar_width,
        label='YOLOv12x', alpha=0.8, color='skyblue')
plt.bar(x_pos_overall, pipeline_overall, bar_width,
        label='Pipeline', alpha=0.8, color='lightcoral')
plt.bar(x_pos_overall + bar_width, yolov8x_overall, bar_width,
        label='YOLOv8x', alpha=0.8, color='lightgreen')

plt.title('Overall Metrics Comparison', fontweight='bold')
plt.ylabel('Score')
plt.xlabel('Metrics')
plt.xticks(x_pos_overall, overall_metrics)
plt.legend(loc='lower right')
plt.grid(True, alpha=0.3)

for container in plt.gca().containers:
    plt.bar_label(container, fmt='%.3f', padding=3, fontsize=8)

plt.tight_layout()
plt.show()
```
